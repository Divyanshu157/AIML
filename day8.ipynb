```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("tips.csv")
df.head()
```

```python
#line _plot using pandas
fig= plt.figure(figsize=(18,6))
df['tip'].plot()
df['total_bill'].plot(color='red',linestyle='-.',marker='o',markerfacecolor='green',markeredgecolor="yellow")

plt.title("Line Chart By PAndas")
plt.grid('on')
plt.legend(['total_bill','tip'])
plt.show()
```

```python
#line plot using seaborn
sns.lineplot(data=df,x=df.index,y='total_bill')
plt.title("Line Chart By Seaborn")
plt.show()
```

```python
df[['total_bill','tip']].plot(color='red')  # this is advantages of pandas over seaborn
```

```python

#Histogram
fig = plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
df['size'].plot.hist(color="yellow",bins=5)
plt.grid()
plt.title('Using pandas')
plt.show()

plt.subplot(1,2,2)
sns.histplot(data=df,x='size',bins=5,color='red')
plt.title('Using Seaborn')
plt.grid()
plt.show()
```

```python
#bar -plot
fig = plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
df['size'].value_counts().plot.bar()
plt.title('Pandas')


plt.subplot(1,2,2)
sns.barplot(data=df,x='day',y='total_bill',estimator=sum)
plt.title('seaborn')
plt.show()
```
