```python
import seaborn as sns
import pandas as pd
from IPython.display import display , Image
```

```python
import matplotlib.pyplot as plt
```

# remaining Graph of day8


```python
df= pd.read_csv('tips.csv')
```

```python

#Boxplot
plt.subplot(1,2,1)
df['total_bill'].plot.box()
plt.title('Pandas')
plt.show()

plt.subplot(1,2,2)
sns.boxplot(data=df,y='total_bill')
plt.title('seaborn')
```

```python
sns.boxenplot(data=df,x='day',y='total_bill')    #pandas cannot do it
```

```python
sns.boxenplot(data=df,x='day',y='total_bill',hue='sex')
```

```python
#scatter plot
plt.figure(figsize=(18,6))
df.plot.scatter(x='total_bill',y='tip')
plt.title('pandas')
plt.show()
plt.subplot(1,2,2)
sns.scatterplot(data=df,x='total_bill',y='tip')
plt.title('Saeborn')
plt.show()
```

```python
sns.scatterplot(data=df,x="total_bill",y='tip',hue='sex',size='size')
plt.show()
```

```python
#hexbin
df.plot.hexbin(x='total_bill',y='tip'),plt.show()
sns.jointplot(data=df,x='total_bill',y='tip',kind='hex'),plt.show()
```

```python
#KDE plot
plt.figure(figsize=(18,6))


plt.subplot(131)
df['total_bill'].plot.hist()
plt.show()


plt.subplot(132)
df['total_bill'].plot.kde()
plt.title('Pandas')
plt.show()

plt.subplot(133)
sns.kdeplot(data=df,x='total_bill',hue='day')
plt.title('seaborn')
plt.show()
```

```python
#Pie chart

# df['total_bill'].plot.pie()
df['day'].value_counts().plot.pie(autopct='%d')

plt.show()

#seaborn  does not have pie function
```

```python
# Voilin Plot 
sns.violinplot(data=df, x='day' , y='tip');plt.show()




#swarm Plot
sns.swarmplot(data=df,x='day',y='tip'); plt.show()
```

```python
#heatmap
sns.heatmap(df[['total_bill','tip','size']]);plt.show()
```

```python
sns.heatmap(df[['total_bill','tip']].corr())
```
