```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

```

```python
iris = datasets.load_iris()
X=iris.data[:,:2]             #considering only two columns

y=iris.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=45)
```

```python
k_list=list(range(1,50,2))
all_score=[]

for k in k_list: 

    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train,y_train)
    y_pred= model.predict(X_test)
    score = accuracy_score(y_test,y_pred)
    all_score.append(score)
     
    
```

```python
all_score
```

```python
plt.plot(k_list,all_score,marker = 'o',linestyle='-')
plt.xlabel("K values")                    # it was not working but as i restart terminal it has started workig properly 
plt.ylabel("Accuracy score")         # restart  if  it does not work

plt.show()
```

```python
best_k=k_list[all_score.index(max(all_score))]
```

```python
best_knn= KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train,y_train)


```

```python

from mlxtend.plotting import plot_decision_regions
plt.figure(figsize=(12,6))
plot_decision_regions(X_train,y_train,clf = best_knn)
plt.ylabel('sepal len')
plt.xlabel("sepal wid")
plt.title('Decision boundary')
plt.show()
```
