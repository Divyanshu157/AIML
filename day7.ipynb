```python
df['total_price']= df['total_bill']+df['tip']
df
```

```python
df.groupby('sex')['total_price'].mean()
```

```python
def mean_Placer(data):

    return df[df['sex']==data['sex']]['total_price'].mean()


```

```python
df['mean_price_sex_wise']=df[['sex','total_price']].apply(mean_Placer,axis=1)
df.head()
```
